## Heart Disease Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
```

```python
heart_disease.isnull().sum()
```

```python
heart_disease["age"].plot.hist(title = "Frequency of Heart Diseases in various age groups",xlabel= "Age");
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
fig,ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
y=over_50["chol"],
c=over_50["target"])

ax.set(title="Cholestrol Levels in various age groups",
xlabel="Age",
ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
plt.style.use('seaborn-whitegrid')

fig,(ax0,ax1) = plt.subplots(nrows=2,
ncols=1,
figsize=(10,10),
)

scatter = ax0.scatter(x=over_50["age"],
y=over_50["chol"],
c=over_50["target"],
cmap="winter"
)

ax0.set(title="Cholestrol levels in age groups above 50",
xlabel="Age",
ylabel="Cholestrol")
ax0.set_xlim([50,80])
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["chol"].mean(),
linestyle="--",
color="red")


scatter2 = ax1.scatter(x=over_50["age"],
y=over_50["thalach"],
c=over_50["target"],
cmap="winter")

ax1.set_xlim([50,80])
ax1.set_ylim([80,200])
ax1.set(title="Max Heart Rate by Age Group",
xlabel="Age",
ylabel="Max Heart Rate")

ax1.legend(*scatter2.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),
linestyle="--",
color="red");


fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
